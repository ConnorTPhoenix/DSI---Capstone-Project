```python
lf['id'] = lf['Spotify Track ID']
sf.columns
```

```python
X = sf.drop(['track_href', 'analysis_url', 'Unnamed: 0', 'uri', 'type'], axis=1)
X.isnull().sum()

# X_scaled = StandardScaler().fit_transform(X)
X_lookup = pd.merge(X,lf[['id', 'Artist_Track']].drop_duplicates(), on ='id')
```

```python
X = X.drop('id', axis=1)
```

```python
from sklearn.preprocessing import Imputer
imp_mean = Imputer(strategy='mean', axis=0)
X_mean = imp_mean.fit_transform(X)
X = pd.DataFrame(X_mean, columns=X.columns)
```

```python
Xs = StandardScaler().fit_transform(X)
```

```python
X = pd.DataFrame(Xs, columns=X.columns)
y = sf['mode']
```

```python
knn = KNeighborsClassifier()
knn.fit(X,y)
```

```python
Artist_Track = 'Jay-Z: Hard Knock Life'
NoSongs = 15
Neighbors = knn.kneighbors(X.iloc[X_lookup[X_lookup['Artist_Track'] == Artist_Track].index.values[0],:], n_neighbors=NoSongs, return_distance=False)
```

```python
kmeans = cluster.KMeans(n_clusters=10)
kmeans.fit(X)
labels = kmeans.labels_
#     print kmeans.cluster_centers_
#     print 'Inertia:', kmeans.inertia_
#     print 'Silhouette:', metrics.silhouette_score(X, labels, metric='euclidean')
```

```python
X['labels'] = labels
```

```python
X.shape
```

```python
X.groupby('labels').mean().plot(kind='bar', figsize=(16,8), width=.8)
plt.legend(bbox_to_anchor=(1.05, 1))
plt.xlabel('Cluster Labels', fontsize=20)
plt.ylabel('Mean Standardized Value', fontsize=20)
plt.title('K-Means Clustering (No. Tracks = 24,323, K=10)', fontsize=24, y=1.02)
plt.xticks(rotation=0, fontsize=16)
```

```python
for c in X.drop('labels',axis=1).columns:
    X.groupby('labels')[c].mean().T.plot(kind='bar', figsize=(20,10), width=.8, color='g')
    plt.title(c, fontsize=24)
    plt.xlabel('Cluster Label', fontsize=20)
    plt.xticks(rotation=0, fontsize=16)
    plt.show()
```
