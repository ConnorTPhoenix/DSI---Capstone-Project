```python
def generate_recommendations(user, no_songs):
    predicts = []
    for song in matrix.index:
        index = cosim_df_read[song].sort_values().iloc[-4:-1].index
        cosim=cosim_df_read[song].sort_values().iloc[-4:-1].values
        predicts.append((cosim[0]*matrix.loc[index[0],user] + cosim[1]*matrix.loc[index[1], user])/(cosim[0]+cosim[1]))
    rf = pd.DataFrame(zip(predicts, matrix[user]))
    rf.index = matrix.index
    rf.columns = ['Recommendations', 'Actual Plays']
    return rf.sort('Recommendations', ascending=False).head(10).head(no_songs)
```

```python
generate_recommendations('user_000577', 10)
```
