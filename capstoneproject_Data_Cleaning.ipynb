## Create track_lookup df, audio_feats df, and Y (popularity) 

```python
sf = sf.drop(['track_href', 'analysis_url', 'Unnamed: 0', 'uri', 'type'], axis=1)
track_lookup = pd.merge(sf,lf[['id', 'Artist_Track']].drop_duplicates(), on ='id')
y = sf['popularity']
```

```python
audio_feats = sf.dropna(axis=1)
audio_feats = sf.drop(['id'], axis=1)
```

```python
from sklearn.preprocessing import Imputer
imp_mean = Imputer(strategy='mean', axis=0)
audio_feats_mean = imp_mean.fit_transform(audio_feats)
audio_feats = pd.DataFrame(audio_feats_mean, columns=audio_feats.columns)
```

```python
audio_feats
```

```python
audio_feats = pd.merge(audio_feats, pd.get_dummies(audio_feats['key'], prefix='k'), right_index=True, left_index=True)
audio_feats = pd.merge(audio_feats, pd.get_dummies(audio_feats['time_signature'], prefix='t'), right_index=True, left_index=True)
```

```python
audio_feats.drop(['key', 'time_signature'], axis=1, inplace=True)
```

```python
audio_feats[[u'energy', u'liveness', u'tempo', u'speechiness',
       u'acousticness', u'instrumentalness', u'danceability', u'duration_ms',
       u'loudness', u'valence', u'mode']]=pd.DataFrame(StandardScaler().fit_transform(audio_feats[[u'energy', u'liveness', u'tempo', u'speechiness',
       u'acousticness', u'instrumentalness', u'danceability', u'duration_ms',
       u'loudness', u'valence', u'mode']]), columns=[u'energy', u'liveness', u'tempo', u'speechiness',
       u'acousticness', u'instrumentalness', u'danceability', u'duration_ms',
       u'loudness', u'valence', u'mode'])
```
