```python
import pandas as pd
import numpy as np
import math
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline
warnings.filterwarnings('ignore')
from sklearn.metrics.pairwise import cosine_similarity
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
matrix = pd.read_csv('../../Dropbox/datasets/matrix.csv')
```

```python
matrix['Total Plays']=matrix.sum(axis=1)
matrix['Total Plays'].describe()
```

```python
# matrix['Total Plays'].hist(bins=200, figsize=(12,8))
# plt.xlim(0, 2500)
sns.set(rc={"figure.figsize": (12, 8)})
sns.distplot(matrix[matrix['Total Plays']>1]['Total Plays'], bins=1000, kde=False, norm_hist=False)
plt.xlim(0,1500)
plt.xlabel('Total Plays', fontsize=18)
plt.yticks(fontsize=14)
plt.xticks(fontsize=14);
```
